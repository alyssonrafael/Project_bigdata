```python
from datetime import datetime

# Suponha que você tenha um DataFrame chamado df com uma coluna 'data_hora' contendo a data e hora completa

# Converta a coluna 'data_hora' para o tipo datetime, se necessário
df['data'] = pd.to_datetime(df['data'])

# Extrai apenas o ano da coluna 'data_hora'
df['ano'] = df['data'].dt.strftime('%Y')

# Imprime o DataFrame com a coluna 'ano' contendo apenas o ano
print(df['ano'])
```

```python

# esse primeiro serve para converter todas a virgulas em pontos as vezes ele da erro i e so descomentar ou comentar
df['subtotal'] = df['subtotal'].str.replace(',', '.')
#fazendo uma conversao para todos da coluna subtotal ficar em float
df['subtotal'] = df['subtotal'].astype(float).round(2)

# Imprime o DataFrame atualizado
print(df['subtotal'])
```

```python
#fazendo a soma geral com objetivo de ver o faturamento a cada ano do dataset
# e tambem da pra pegar a quantidade de pedidos e de cancelamento

df_ano=df.groupby('ano')

soma_por_ano = df_ano.sum()

soma_por_ano['subtotal']
```
